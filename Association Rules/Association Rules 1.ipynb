```python
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
book=pd.read_csv('/users/rahulkp/downloads/book.csv')
book
```

```python
# Apriori Algorithm
frequent_items=apriori(book,min_support=0.05,use_colnames=True)
frequent_items
```

```python
rules=association_rules(frequent_items,metric='lift',min_threshold=0.4)
rules
```

```python
# with 80% confidence
rules=association_rules(frequent_items,metric='lift',min_threshold=0.8)
rules
```

```python
rules.sort_values('lift',ascending=False)
```

An leverage value of 0 indicates independence. Range will be [-1 to 1]

A high conviction value means that the consequent is highly depending on the antecedent and range [0 to inf]

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
import matplotlib.pyplot as plt
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

# Association rules with 10% Support and 70% confidence
